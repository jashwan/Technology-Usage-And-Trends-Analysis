### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
##Write your code here
df_filtered = df.dropna(subset=['JobSatPoints_6', 'JobSatPoints_7'])

# Group by Age and calculate median values for JobSatPoints_6 and JobSatPoints_7
age_grouped = df_filtered.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Plotting the stacked bar chart
plt.figure(figsize=(12, 8))
age_grouped.plot(kind='bar', stacked=True, color=['skyblue', 'lightgreen'])

# Adding labels and title
plt.title('Stacked Bar Chart of Job Satisfaction Across Age Groups', fontsize=14)
plt.xlabel('Age Groups', fontsize=12)
plt.ylabel('Median Job Satisfaction Score', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

##### Stacked Chart of `JobSatPoints_6` and `JobSatPoints_7` for Employment Status


Create a stacked chart to compare job satisfaction (`JobSatPoints_6` and `JobSatPoints_7`) across different employment statuses. This will show how satisfaction varies by employment type.


```python
##Write your code here
df_filtered = df.dropna(subset=['JobSatPoints_6', 'JobSatPoints_7'])

# Group by Employment status and calculate median values for JobSatPoints_6 and JobSatPoints_7
employment_grouped = df_filtered.groupby('Employment')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Plotting the stacked bar chart
plt.figure(figsize=(12, 8))
employment_grouped.plot(kind='bar', stacked=True, color=['skyblue', 'lightgreen'])

# Adding labels and title
plt.title('Stacked Bar Chart of Job Satisfaction Across Employment Status', fontsize=14)
plt.xlabel('Employment Status', fontsize=12)
plt.ylabel('Median Job Satisfaction Score', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

### Task 2: Stacked Chart for Compensation and Job Satisfaction by Age Group


##### This stacked chart visualizes the composition of compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) specifically for respondents aged 30-35.


```python
##Write your code here
# Filter data for respondents aged 30-35
age_group_30_35 = df[df['Age'].isin(['30-34 years old', '35-44 years old'])]

# Drop rows with missing values in 'ConvertedCompYearly' and 'JobSatPoints_6'
age_group_30_35 = age_group_30_35.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6'])

# Calculate the median for ConvertedCompYearly and JobSatPoints_6 for this age group
median_compensation = age_group_30_35['ConvertedCompYearly'].median()
median_job_satisfaction = age_group_30_35['JobSatPoints_6'].median()

# Prepare data for the stacked bar chart
data = {
    'Category': ['Compensation', 'Job Satisfaction'],
    'Median Value': [median_compensation, median_job_satisfaction]
}

# Create a DataFrame for plotting
df_stacked = pd.DataFrame(data)

# Plotting the stacked bar chart
plt.figure(figsize=(8, 6))
plt.bar(df_stacked['Category'], df_stacked['Median Value'], color=['lightblue', 'lightgreen'])

# Adding labels and title
plt.title('Stacked Bar Chart of Compensation and Job Satisfaction for Age Group 30-35', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Median Value', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

##### Stacked Chart of Median Compensation and Job Satisfaction Across Age Group


Compare the median compensation and job satisfaction metrics across different age groups. This helps visualize how compensation and satisfaction levels differ by age.


```python
##Write your code here
df_filtered = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6'])

# Group the data by age group and calculate the median for 'ConvertedCompYearly' and 'JobSatPoints_6'
median_values = df_filtered.groupby('Age').agg({
    'ConvertedCompYearly': 'median',
    'JobSatPoints_6': 'median'
}).reset_index()

# Plotting the stacked bar chart
plt.figure(figsize=(10, 6))

# Stacking the bars for compensation and job satisfaction
plt.bar(median_values['Age'], median_values['ConvertedCompYearly'], label='Compensation', color='lightblue')
plt.bar(median_values['Age'], median_values['JobSatPoints_6'], label='Job Satisfaction', color='lightgreen', bottom=median_values['ConvertedCompYearly'])

# Adding labels and title
plt.title('Stacked Bar Chart of Median Compensation and Job Satisfaction Across Age Groups', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Median Value', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Metrics", fontsize=10)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
##Write your code here
df_filtered = df.dropna(subset=['Age', 'DatabaseWantToWorkWith'])

# Explode the 'DatabaseWantToWorkWith' column to split multiple databases into individual rows
df_exploded = df_filtered.set_index('Age')['DatabaseWantToWorkWith'].str.split(';', expand=True).stack().reset_index(name='DatabaseWantToWorkWith')

# Group by age and database, and get the count of each combination
age_db_counts = df_exploded.groupby(['Age', 'DatabaseWantToWorkWith']).size().unstack(fill_value=0)

# Plotting the stacked bar chart
plt.figure(figsize=(12, 8))

# Plot the stacked chart for the databases
age_db_counts.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='tab20')

# Adding labels and title
plt.title('Stacked Chart of Preferred Databases by Age Group', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Databases", fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
##Write your code here
# Drop rows with missing values in 'JobSat' and 'Employment'
df_filtered = df.dropna(subset=['JobSat', 'Employment'])

# Group by Job Satisfaction level and Employment type, then count the occurrences
job_sat_emp_type_counts = df_filtered.groupby(['JobSat', 'Employment']).size().unstack(fill_value=0)

# Plotting the stacked bar chart
plt.figure(figsize=(12, 8))

# Plot the stacked chart for Employment types across Job Satisfaction levels
job_sat_emp_type_counts.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='Set3')

# Adding labels and title
plt.title('Stacked Chart of Employment Type by Job Satisfaction', fontsize=14)
plt.xlabel('Job Satisfaction Level', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Employment Type", fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

### Task 4: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
df['Age'].unique()
```

```python
##Write your code here
# Define a mapping to convert text age categories into numeric values
age_group_mapping = {
    'Under 18 years old': 0,
    '18-24 years old': 1,
    '25-34 years old': 2,
    '35-44 years old': 3,
    '45-54 years old': 4,
    '55-64 years old': 5,
    '65 years or older': 6,
    'Prefer not to say': 7
}

# Replace age categories with numeric values based on the mapping
df['AgeGroupNumeric'] = df['Age'].map(age_group_mapping)

# Optionally, drop rows with 'Prefer not to say' or handle it differently if you want
df_filtered = df[df['AgeGroupNumeric'] != 7]

# Now, we will proceed with the stacked chart creation based on the numeric age groups

# Split LanguageAdmired values and normalize data
df_filtered['LanguageAdmired'] = df_filtered['LanguageAdmired'].str.split(';')

# Explode the list of languages into separate rows
df_exploded = df_filtered.explode('LanguageAdmired')

# Group by Age Group and Language Admired, then count occurrences
language_counts = df_exploded.groupby(['AgeGroupNumeric', 'LanguageAdmired']).size().unstack(fill_value=0)

# Plotting the stacked chart
plt.figure(figsize=(12, 8))

# Create a stacked bar chart for programming languages by age group
language_counts.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='Set3')

# Adding labels and title
plt.title('Stacked Chart for Preferred Programming Languages by Age Group', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)
plt.xticks(ticks=range(len(language_counts.index)), labels=['Under 18', '18-24', '25-34', '35-44', '45-54', '55-64', '65+'], rotation=45)
plt.legend(title="Programming Language", fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

##### 2. Stacked Chart for Technology Adoption by Employment Type


Explore how admired platforms (`PlatformAdmired`) differ across employment types (e.g., full-time, freelance)


```python
##Write your code here
# Split the 'PlatformAdmired' values and normalize data
df['PlatformAdmired'] = df['PlatformAdmired'].str.split(';')

# Explode the list of platforms into separate rows
df_exploded = df.explode('PlatformAdmired')

# Group by Employment Type and Platform Admired, then count occurrences
platform_counts = df_exploded.groupby(['Employment', 'PlatformAdmired']).size().unstack(fill_value=0)

# Plotting the stacked chart
plt.figure(figsize=(12, 8))

# Create a stacked bar chart for platforms admired by employment type
platform_counts.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='Set3')

# Adding labels and title
plt.title('Stacked Chart for Technology Adoption by Employment Type', fontsize=14)
plt.xlabel('Employment Type', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title="Admired Platform", fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
