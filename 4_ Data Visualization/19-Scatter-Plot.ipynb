# **Scatter Plot**


## Objectives




- Create and analyze scatter plots to examine relationships between variables.

- Use scatter plots to identify trends and patterns in the dataset.

- Focus on visualizations centered on scatter plots for better data-driven insights.


## Setup: Working with the Database



**Install and import the required libraries**


```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Step 1: Load the dataset


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv(file_path)
```


### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
## Write your code here
age_map = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70,
    'Prefer not to say': None
}

# Apply mapping
df['AgeNum'] = df['Age'].map(age_map)

# Drop missing values
df_scatter = df[['AgeNum', 'JobSatPoints_6']].dropna()

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(df_scatter['AgeNum'], df_scatter['JobSatPoints_6'], alpha=0.5, color='teal')
plt.title('Scatter Plot of Age vs. Job Satisfaction')
plt.xlabel('Age (Estimated)')
plt.ylabel('Job Satisfaction Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Compensation vs. Job Satisfaction


Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
## Write your code here
df_scatter = df[['ConvertedCompYearly', 'JobSatPoints_6']].dropna()

# Optional: Filter extreme compensation outliers for better visualization
df_scatter = df_scatter[df_scatter['ConvertedCompYearly'] < df_scatter['ConvertedCompYearly'].quantile(0.99)]

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(df_scatter['ConvertedCompYearly'], df_scatter['JobSatPoints_6'], alpha=0.5, color='purple')
plt.title('Scatter Plot of Compensation vs. Job Satisfaction')
plt.xlabel('Yearly Compensation (ConvertedCompYearly)')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
## Write your code here
import seaborn as sns
df_scatter_age = df[['Age', 'JobSatPoints_6']].dropna()

# Convert 'Age' to numeric values
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 65,
    'Prefer not to say': 50  # Or another placeholder, depending on the dataset's distribution
}

df_scatter_age['Age'] = df_scatter_age['Age'].map(age_mapping)

# Plotting with seaborn regplot for trend line
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='JobSatPoints_6', data=df_scatter_age, scatter_kws={'alpha':0.5}, line_kws={'color': 'red'})
plt.title('Scatter Plot of Age vs. Job Satisfaction with Trend Line')
plt.xlabel('Age')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
## Write your code here
df_scatter_age_workexp = df[['Age', 'YearsCodePro']].dropna()

# Convert 'Age' to numeric values
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 65,
    'Prefer not to say': 50  # Or another placeholder, depending on the dataset's distribution
}

df_scatter_age_workexp['Age'] = df_scatter_age_workexp['Age'].map(age_mapping)

# Convert 'YearsCodePro' to numeric values
df_scatter_age_workexp['YearsCodePro'] = pd.to_numeric(df_scatter_age_workexp['YearsCodePro'], errors='coerce')

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='YearsCodePro', data=df_scatter_age_workexp, alpha=0.5)
plt.title('Scatter Plot of Age vs. Work Experience (YearsCodePro)')
plt.xlabel('Age')
plt.ylabel('Years of Professional Coding Experience (YearsCodePro)')
plt.tight_layout()
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
## Write your code here
df_bubble = df[['ConvertedCompYearly', 'JobSatPoints_6', 'Age']].dropna()

# Plotting the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(
    x=df_bubble['ConvertedCompYearly'],
    y=df_bubble['JobSatPoints_6'],
    s=df_bubble['Age'] * 10,  # Multiply by 10 to adjust bubble size
    alpha=0.5,  # Transparency to avoid overlapping bubbles
    c=df_bubble['Age'],  # Color based on Age
    cmap='viridis',  # Color map for visual distinction
)

# Add titles and labels
plt.title('Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size')
plt.xlabel('Yearly Compensation (ConvertedCompYearly)')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.colorbar(label='Age')  # Color bar to represent age values
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here
df_scatter = df[['LanguageHaveWorkedWith', 'JobSatPoints_6']].dropna()

# Calculate the average job satisfaction for each programming language
language_satisfaction = df_scatter.groupby('LanguageHaveWorkedWith')['JobSatPoints_6'].mean().reset_index()

# Sort the data by average job satisfaction
language_satisfaction = language_satisfaction.sort_values(by='JobSatPoints_6', ascending=False)

# Plotting the scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=language_satisfaction,
    x='LanguageHaveWorkedWith',
    y='JobSatPoints_6',
    size='JobSatPoints_6',  # Use JobSat as the size of the points
    sizes=(50, 200),  # Adjust point size range
    hue='JobSatPoints_6',  # Color points by JobSat for better visual clarity
    palette='viridis',
    legend=None
)

plt.title('Scatter Plot of Popular Programming Languages by Job Satisfaction')
plt.xlabel('Programming Language')
plt.ylabel('Average Job Satisfaction')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here
df_scatter = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'Employment'])

# Plotting the scatter plot with Employment type as hue
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df_scatter,
    x='ConvertedCompYearly',
    y='JobSatPoints_6',
    hue='Employment',  # Differentiate by Employment type
    style='Employment',  # Different marker styles for each employment type
    palette='Set1',  # Color palette for clear differentiation
    s=100,  # Marker size
    alpha=0.7  # Transparency of the points
)

plt.title('Scatter Plot of Compensation vs. Job Satisfaction by Employment Type')
plt.xlabel('Yearly Compensation')
plt.ylabel('Job Satisfaction')
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
## Write your code here
df_scatter_age_country = df.dropna(subset=['YearsCodePro', 'Age', 'Country'])

# Convert YearsCodePro to numeric (for plotting purposes)
df_scatter_age_country['YearsCodePro'] = pd.to_numeric(df_scatter_age_country['YearsCodePro'], errors='coerce')

# Plotting the scatter plot
plt.figure(figsize=(14, 8))
sns.scatterplot(
    data=df_scatter_age_country,
    x='YearsCodePro',
    y='Age',
    hue='Country',  # Color by Country
    style='Age',  # Different marker styles for each Age group
    palette='tab20',  # Color palette for clear differentiation
    s=100,  # Marker size
    alpha=0.7  # Transparency of the points
)

plt.title('Scatter Plot for Work Experience vs. Age Group by Country')
plt.xlabel('Years of Work Experience (YearsCodePro)')
plt.ylabel('Age Group')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
