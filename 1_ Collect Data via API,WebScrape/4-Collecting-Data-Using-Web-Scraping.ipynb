## Extract information from web site
You will extract the data from web site: <br>


```python

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before start to scrape.


Import the required libraries


```python

from bs4 import BeautifulSoup
import pandas as pd
import requests
```

Download the webpage at the url


```python

data=requests.get(url).text
```

Create a soup object


```python

soup=BeautifulSoup(data,'html.parser')
```

Scrape the `Language name` and `annual average salary`.


```python

table=soup.find('table')
program_language=[]
for row in table.find_all('tr'):
    column=row.find_all('td')
    language=column[1].getText()
    avg_salary=column[3].getText()
    program_language.append(str(language)+','+ str(avg_salary).replace('$','').replace(',',''))
program_language
```

Save the scrapped data into a file named *popular-languages.csv*


```python

df=pd.DataFrame(program_language)
df.to_csv('popular-languages.csv',header=None,index=None)
```
